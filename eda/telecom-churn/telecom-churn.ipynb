## 1. Telecom Churn - Exploratory Data Analysis Using Pandas

```python
# Import Necessary Headers
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read CSV File
df = pd.read_csv("telecom_churn.csv")
df.head()
```

```python
print(df.shape)
```

```python
print(df.info())
```

- we can observe that the there are no null values in the dataset since the `Non-Null Count` mathes the shape of the df. We can also cross verify using the `isnull.sum()` method

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
df.describe(include=["object", "bool"])
```

- The "Churn" column is a boolean (`bool`), which we need to convert to an integer (0 and 1).  
- This helps with compatibility in machine learning models.  

```python
df["Churn"] = df["Churn"].astype("int64")
df.head()
```

```python
print(df["Churn"].value_counts(normalize = True) * 100)

plt.figure(figsize=(16,4))
sns.countplot(x="Churn", data=df)
plt.title("Churn Distribution")
plt.xticks([0, 1], ["False", "True"]) 
plt.show()
```

- we can observe that there is a massive imbalance in the dataset
